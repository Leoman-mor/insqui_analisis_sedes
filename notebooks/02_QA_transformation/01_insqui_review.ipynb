```python
import pandas as pd
from branca.element import Figure
import folium
```

## Carga de datos extraidos de la página del INSQUI

```python
df = pd.read_csv('../../data/raw/insqui/coordenadas_insqui.csv', header=0)

df[['tipo_actividad', 'latitud', 'longitud']] = (
    df.iloc[:, 0].str.split(';', expand=True)
)

df['latitud'] = df['latitud'].astype(float)
df['longitud'] = df['longitud'].astype(float)

df = df.drop(columns=['tipo_actividad;latitud;longitud'])
```

```python
df.describe()
```

## Graficación datos iniciales

```python
fig = Figure(width=1000, height=700)

mapa_inical = folium.Map(
    location=[4.6, -74.1],
    zoom_start=6,
    tiles="CartoDB positron"
)

for _, r in df.iterrows():
    folium.CircleMarker(
        location=[r["latitud"], r["longitud"]],
        radius=4,
        popup=r["tipo_actividad"],
        fill=True,
        fill_opacity=0.7
    ).add_to(mapa_inical)

fig.add_child(mapa_inical)
fig
```

## Guardar df

```python
df.to_csv('../../data/transformed/insqui/insqui_coordinates_clean.csv', index=False)
```
